**NOW WE HAVE TO DO THIS FOR ALL THE PAGES**

```python
new = pd.DataFrame()
```

```python
for i in range(1, 429):
  response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={}'.format(i))
  df = pd.DataFrame(response.json()['results'])[['title', 'overview', 'genre_ids']]
  new = new.append(df, ignore_index = True)
```

```python
new
```

```python
new.dropna(subset = ['genre_ids'], inplace = True , axis = 0)
```

```python
new.reset_index(drop = True, inplace = True)
```

```python
new
```

```python
df = pd.DataFrame(new)
replace = {'28' : 'Action', '14':"Fantasy", '36':"History", '27':"Horror", '10402':"Music",
          '12': 'Adventure','9648':"Mystery", '10749':"Romance", '878':"Science Fiction",'10770':"TV Movie",
           '53':"Thriller",'10752':"War", '37':"Western",
           '18': 'Drama',
           '16': 'Animation',
           '35': 'Comedy',
           '80': 'Crime',
           '99':"Documentary",
           '10751':"Family",
           }
for i in range(len(df)):
    for j in range(len(df['genre_ids'][i])):
        if str(df['genre_ids'][i][j]) in replace:
            df['genre_ids'][i][j] = replace[str(df['genre_ids'][i][j])]
```
